## How to Write JSON to a File
In the first example we are going to use the Python module json to write JSON to a file. It's very simple, we just need to import json. All we need to do is to store our data in a dict object. This is the code for the tutorial [How to Read & Write JSON files using Python and Pandas](https://www.marsja.se/how-to-read-and-write-json-files-using-python-and-pandas/)

```python
data = {"Sub_ID":["1","2","3","4","5","6","7","8" ],
        "Name":["Erik", "Daniel", "Michael", "Sven",
                "Gary", "Carol","Lisa", "Elisabeth" ],
        "Salary":["723.3", "515.2", "621", "731", 
                  "844.15","558", "642.8", "732.5" ],
        "StartDate":[ "1/1/2011", "7/23/2013", "12/15/2011",
                     "6/11/2013", "3/27/2011","5/21/2012", 
                     "7/30/2013", "6/17/2014"],
        "Department":[ "IT", "Manegement", "IT", "HR", 
                      "Finance", "IT", "Manegement", "IT"],
        "Sex":[ "M", "M", "M", 
              "M", "M", "F", "F", "F"]}

print(data)
```

```python
# Saving to a JSON file
import json

with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```

```python
# Open the file using json
with open('data.json') as json_file:
    data = json.load(json_file)

print(data)
```

## How to Load Data from JSON using Pandas
In the next example we are going to use Pandas read_json method to read the JSON file we wrote earlier (i.e., data.txt). It's fairly simple we start by importing pandas as pd:

```python
import pandas as pd

df = pd.read_json('data.json')

df
```

## Data Manipulation using Pandas
Here we are going to use Pandas inplace method to modify our dataframe. We start by setting the Sub_ID column as index. 

```python
df.set_index('Sub_ID', inplace=True)
df
```

# Pandas JSON to CSV
If we want to save the JSON file to a csv file this is also possible. We just use the to_csv method from Pandas to save the dataframe to a csv file.

```python
df.to_csv('data.csv')
```

### Load JSON from an URL Second Example
When loading some data from JSON, using Pandas read_json seems to create a dataframe with dictionaries within each cell. One way to deal with these dictionaries, nested within dictionaries, is to work with the Python module request.

```python
url = "https://think.cs.vt.edu/corgis/json/airlines/airlines.json"
df = pd.read_json(url)
df.head()
```

```python
import requests
from pandas.io.json import json_normalize

url = "https://think.cs.vt.edu/corgis/json/airlines/airlines.json"
resp = requests.get(url=url)

df = json_normalize(resp.json())
df.head()
```

### Cleaning up the column names
As can be seen in the image above, the column names are quite long. This is quite impractical when we are going to create a time series plot, later, using Seaborn.

```python
import re

df.rename(columns=lambda x: re.sub("statistics.# of","",x), 
          inplace=True)
df.rename(columns=lambda x: re.sub("statistics.","",x), 
          inplace=True)

df.columns = df.columns.str.replace("[.]", "_")
df.head()
```

### Time Series Plot
Plotting the data obtained from JSON

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(10, 7))
g = sns.lineplot(x="timeyear", y="flightscancelled", ci=False,
             hue="carriercode", data=df)

g.set_ylabel("Flights Cancelled",fontsize=20)
g.set_xlabel("Year",fontsize=20)


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

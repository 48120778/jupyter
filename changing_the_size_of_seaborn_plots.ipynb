### Changing the Font Size on a Seaborn Plot
As can be seen, the fonts are now small. We can of course change the fons using the set method and the font_scale argument. Again, we are going to use the iris dataset so we may need to load it again. Furthermore, we are changing the size of the scatter plot using Matplotlib Pyplot, gcf(), and the set_size_inches() method:

```python
iris_df = sns.load_dataset('iris')

fig = plt.gcf()
fig.set_size_inches(12, 8)

# Setting the font scale
sns.set(font_scale=2)
sns.scatterplot(x='sepal_length', 
                y='petal_length', 
                data=iris_df)

plt.savefig(plotd + 'scatter_plot_in_Python_change_size_and_font_size.jpg', format='jpeg', dpi=70)
```

## Changing Size on Subplots

```python
sns.set(font_scale=1)

g = sns.FacetGrid(iris_df, col="species")
g = g.map(plt.scatter, "petal_length", "sepal_length")
```

```python
sns.set(style="ticks")
g = sns.FacetGrid(iris_df, col="species")
g = g.map(plt.scatter, "petal_length", "petal_width")
g.fig.set_figheight(6)
g.fig.set_figwidth(10)

plt.savefig(plotd + 'how_to_save_a_scatter_plot_as_jpeg.jpg', format='jpeg', dpi=70)
```
